```python
# Importing Dependencies

import pandas as pd
import os
```

```python
# Loading in the vine_table csv file

file_to_load = os.path.join('vine_table.csv')
```

```python
# Reading the csv file into a Pandas DataFrame:

vine_df = pd.read_csv(file_to_load)

vine_df.head()
```

```python
# filtering the vine_df for rows with total_votes >= 20

twenty_votes = vine_df[vine_df['total_votes'] >= 20]

twenty_votes
```

```python
# filtering the above DataFrame for reviews where the "helpful_votes" divided by "total_votes" >= 50%


fifty_percent = twenty_votes.loc[(twenty_votes['helpful_votes']/ twenty_votes['total_votes']) >= 0.5]

fifty_percent
```

```python
# Isolating all of the rows of data where a review was written as a part of the Vine program

vine_yes = fifty_percent[fifty_percent['vine'] == 'Y']

vine_yes
```

```python
# Isolating all of the rows of data where a review written was not part of the Vine program.

vine_no = fifty_percent[fifty_percent['vine'] == 'N']

vine_no
```

```python
# Calculating the percentage of 5 star reviews among vine subscribers

total_rev_vine = vine_yes['review_id'].count()

vine_five = vine_yes.loc[(vine_yes['star_rating']==5.0)]['review_id'].count()

percentage_five_vine = (vine_five/total_rev_vine) * 100

percentage_five_vine
```

```python
# Calculating the percentage of 5 star reviews among non-vine subscribers

total_rev_no_v = vine_no['review_id'].count()

no_vine_5 = vine_no.review_id[vine_no['star_rating']==5.0].count()

percentage_5_no_vine = (no_vine_5/total_rev_no_v)*100

percentage_5_no_vine
```

```python
summary = (
f"--------------\n"
f"Vine Subscribers\n"
f"--------------\n"
f"Total Reviews from Vine Subscribers: {total_rev_vine}\n"
f"Total 5-Star Reviews from Vine Subscribers: {vine_five}\n"
f"Percentage of 5-Star Reviews from Vine Subscribers: {percentage_five_vine:.2f}%\n"
f"--------------\n"
f"Non-Vine Subscribers\n"
f"--------------\n"
f"Total Reviews from Non-Vine Subscribers: {total_rev_no_v}\n"
f"Total 5-Star Reviews from Vine Subscribers: {no_vine_5}\n"
f"Percentage of 5-Star Reviews from Vine Subscribers: {percentage_5_no_vine:.2f}%\n")

print(summary)
```
